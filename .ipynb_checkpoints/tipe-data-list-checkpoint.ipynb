# Tipe Data List

Tipe data list adalah tipe data yang berbentuk daftar yang dipisahkan dengan koma. Contoh [1, 2]

## String dalam list

```python
data = ["Python", "Javascript", "Java", "Kotlin"]

print(data)
print(type(data))
```

## Integer dan Float dalam List

```python
data = [22, 33, 12, 24, 34, 44.5, 10.2, 22.22]

print(data)
print(type(data))
```

## Tuple dalam List

```python
data = [(1, 2, 3), (3, 4, 5), (6, 7, 8)]

print(data)
print(type(data))
```

## Boolean dalam List

```python
data = [True, False, False, True]

print(data)
print(type(data))
```

## Multiple data type List

```python
ganda = [True, '2', 3.0, 4]

print(ganda)
print(type(ganda))
```

# Mengurutkan data dalam List

```python
nilai = [2, 1, 3, 4, 0]

print(nilai)

# Gunakan method sort()

nilai.sort()

print(nilai)
```

```python
test = [1, 2, True, "nama", (2, 3, 2)]

print(test)

test.sort()

print(test)
# Akan menghasilkan error, jadi untuk mengurutkan data dalam tipe data list,
# datanya harus bertipe data integer atau float
```

## Menambahkan data kedalam list

Untuk menambahkan data dalam list kita menggunakan method append()

```python
nama = ['Randang', 'Kalio', 'Gulai']

print(nama)

nama.append('Touco') # Tambahkan nilai

print(nama)
```

```python
thislist = ["apple", "banana", "cherry", "apple", "cherry"]

print(thislist)
```

```python
umur = [16, 18, 20, 16, 21, 21]

print(umur)

umur.append(18)
print(umur)
```

## Mengakses data

```python
data = [1, 2, 3, 4]

print(len(data)) # Mengecek berapa banyak data di list

print(data)
```

```python
nama = ["ayu", "bayu", "cinta", "dinda"]

print(nama[2])
print(nama[3])
```

```python
nama = ["ayu", "bayu", "cinta"]

print(nama[3])
```

```python
data = [1, 2, 3, 4]
print(data[0])
print(data[1])
print(data[2])
print(data[3])
```

```python
nama = ['Python', 'C++', 'PHP']

print(nama[0])
print(type(nama))
print(type(nama[0]))
```

## Reverse List

Reverse adalah membalikan data list

```python
buah = ["anggur", "durian", "jambu"]
print(buah)

buah.reverse()
print(buah)
```

## Extend List
Menggabungkan dua buah list

```python
data1 = [0, 1, 2, 3]
data2 = [4, 5, 6, 7]

print(data1)
print(data2)
print()

data2.extend(data1)
print(data2)
```

1. buatlah type data list dengan valuenya
    nama-nama teman tercinta anda
2. tambahankan data baru didalam list
    tersebut dengan orang yang duduk disamping anda
3. buatlah type data list dengan value masing-masing
    umur dari teman-teman anda
4. urutkan data list tersebut
5. tentukan list index dari data tersebut
6. gabungkan kedua data tersebut

```python
teman = ["ayu", "bayu", "cinta", "dinda"]
print(teman)
teman.append("elmi")
print(teman)
print()

umur = [19, 18, 16, 20, 22]
print(umur)
umur.sort()
print(umur)
print(umur[3])
print()

teman.extend(umur)
print(teman)
```
